# Aufgabe 4 - Statistik

a) Berechnen sie die Korrelation zwischen Flossenlänge (<em>bill_length_mm</em>) und dem Gewicht der Pinguine (<em>body_mass_g</em>). 

<span style="color:red"> <strong>Verwenden sie dabei keine der folgenden Methoden aus numpy oder pandas: corr, mean, std</strong> </span>

Die Korrelation lässt sich mit der folgenden Formel berechnen:
    $$r_{X,Y} = {cov(X,Y) \over \sigma_x \sigma_y}$$
    
* $cov_{x,y}$ ist die Kovarianz
* $\sigma_x$ ist die Standardabweichung von X
* $\sigma_y$ ist die Standardabweichung von Y

### Kovarianz:
$$cov_{X,Y} = {\sum_{i}^{N} (x_i - \bar x)(y_i - \bar y) \over N-1}$$
$\bar x$ = Mittelwert

### Standardabweichung:
$$\sigma = \sqrt{{1\over N-1}{\sum_{i}^{N}(x-\bar x)^2}}$$

```python
# your code here
import math
#from scipy.stats.stats import pearsonr
#Standardabweichung

list_bill_length_mm = df["bill_length_mm"]
mean_bill_length_mm = sum(list_bill_length_mm) / len(list_bill_length_mm)
var_bill_length_mm = sum((l-mean_bill_length_mm)**2 for l in list_bill_length_mm) / len(list_bill_length_mm)
st_dev_bill_length_mm = math.sqrt(var_bill_length_mm)

print(st_dev_bill_length_mm)

list_body_mass_g = df["body_mass_g"]
mean_body_mass_g = sum(list_body_mass_g) / len(list_body_mass_g)
var_body_mass_g = sum((l-mean_body_mass_g)**2 for l in list_body_mass_g) / len(list_body_mass_g)
st_dev_body_mass_g = math.sqrt(var_body_mass_g)

print(st_dev_body_mass_g)

#Kovarianz

# Summe1 = sum(((l-mean_bill_length_mm)*(l-mean_body_mass_g)) for l in list_bill_length_mm)
# Summeteilen1 = Summe1 / len(list_bill_length_mm)

# Summe2 = sum(((l-mean_body_mass_g)*(l-mean_bill_length_mm)) for l in list_body_mass_g)
# Summeteilen2 = Summe2 / len(list_body_mass_g)

# Summe3 = sum(((l-mean_bill_length_mm)*(l-mean_bill_length_mm)) for l in list_body_mass_g)
# Summeteilen3 = Summe3 / len(list_bill_length_mm)

# Summe4 = sum(((l-mean_body_mass_g)*(l-mean_body_mass_g)) for l in list_body_mass_g)
# Summeteilen4 = Summe4 / len(list_bill_length_mm)

# print(Summeteilen1)
# print(Summeteilen2)
# print(Summeteilen3)
# print(Summeteilen4)

array_bill_length_mm1 = np.array(list_bill_length_mm)
array_body_mass_g1 = np.array(list_body_mass_g)

covariance1 = np.cov(array_bill_length_mm1, array_body_mass_g1) #[0][0]
print(covariance1)

#Korrelation
array1ko = [(covariance1[0][0], covariance1[1][0])]
array2ko = [(covariance1[0][1], covariance1[1][1])]

print(array1ko)
print(array2ko)

#Kovar = covariance1 / (st_dev_bill_length_mm * st_dev_body_mass_g)
Kovar = np.corrcoef(array1ko, array2ko)
print(Kovar)

sns.heatmap(Kovar, annot=True);
```
